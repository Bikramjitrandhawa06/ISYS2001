## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
I asked the AI to help me write a simple Python program for tracking a
monthly budget. I told it that the program should ask the user to enter
their monthly income and then ask for three expenses. After collecting
this information, the program should add the expenses together and
calculate how much money is left from the income. Finally, it should
show the remaining budget and give a warning message if the expenses
are more than the income.
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# Ask the user for their monthly income
income = float(input("Please enter your monthly income: "))

# Collect expenses from the user
print("Enter your three monthly expenses:")

expense1 = float(input("Expense 1: "))
expense2 = float(input("Expense 2: "))
expense3 = float(input("Expense 3: "))

# Calculate total expenses
total = expense1 + expense2 + expense3

# Calculate remaining budget
balance = income - total

# Show results
print("\nYour total expenses are:", total)
print("Your remaining budget is:", balance)

# Check if the budget is negative
if balance < 0:
    print("⚠ Warning: You have spent more than your income.")
else:
    print("You are managing your budget well.")
```

### 🐛 Debugging and Refinement

```python
# Document any issues and fixes
issues_found = [
    "Issue 1: At first my program was not calculating the budget correctly.",
    "Fix 1: I realised the numbers from input were treated as text, so I used float() to convert them into numbers.",

    "Issue 2: I forgot to add a warning when the expenses were more than the income.",
    "Fix 2: I added an if condition to check if the remaining budget is below zero and show a warning message."
]

print("🔧 DEBUGGING NOTES:")
for item in issues_found:
    print(f"- {item}")
```

```python
# REFINED VERSION - Your Working Budget Tracker
income = float(input("Monthly income ($): "))

expense1 = float(input("Expense 1 ($): "))
expense2 = float(input("Expense 2 ($): "))
expense3 = float(input("Expense 3 ($): "))

total_expenses = expense1 + expense2 + expense3
remaining = income - total_expenses

print("\n----- Budget Summary -----")
print("Income:          $", round(income, 2))
print("Total expenses:  $", round(total_expenses, 2))
print("Money left:      $", round(remaining, 2))

if remaining < 0:
    print("⚠ You are over budget this month.")
else:
    print("✅ You are within your budget. Nice!")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: Input validation so the program does not accept negative numbers."

enhancement_prompt = """
I asked the AI to help me improve my budget tracker by making sure the
user cannot enter negative values. I wanted the program to check the
numbers entered for income and expenses and ask the user to type the
value again if it is less than zero. I also asked the AI to keep the
solution simple so that it is easy for a beginner to understand.
"""

print(f"🚀 ENHANCEMENT: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
def get_valid_amount(message):
    amount = float(input(message))
    while amount < 0:
        print("Oops! Please enter a value that is 0 or more (no negatives).")
        amount = float(input(message))
    return amount

# Income
income = get_valid_amount("Enter your monthly income ($): ")

# Expenses (up to 3)
print("\nNow enter your three expenses:")
expense1 = get_valid_amount("Expense 1 ($): ")
expense2 = get_valid_amount("Expense 2 ($): ")
expense3 = get_valid_amount("Expense 3 ($): ")

# Calculations
total_expenses = expense1 + expense2 + expense3
remaining_budget = income - total_expenses

# Output
print("\n------ Budget Results ------")
print("Income:            $", round(income, 2))
print("Total expenses:    $", round(total_expenses, 2))
print("Remaining budget:  $", round(remaining_budget, 2))

# Warning if negative
if remaining_budget < 0:
    print("Warning: You are over budget this month!")
else:
    print("You are within your budget. Keep it up!")
```
